```python
class Point():
    def __init__(self,x ,y):
        self.x = x
        self.y = y

blank = Point(3,4)     #  هر دفعه نام کلاس فراخوانی و یک شیء ایجاد شود تابع سازنده یکبار اجرا می شود    
```

```python
class Rectangle():
    def __init__(self,width ,height,corner):
        self.color = "green" # جزو فهرست پارامترها نیست و معمولا برای بیرون از کلاس آن را استفاده نمی کنیم و توابع درونی امکان تغیر آن را دارند
        self.width = width
        self.height = height
        self.corner = Point(corner.x,corner.y)
```

```python
rect = Rectangle(100,200,blank)

print (rect.corner.x , rect.corner.y)
```

```python
#مثال سوم تعریف کلاس تایم 
# متغیرهای کلاس در ساختار کلاس تعریف می شوند 
#  از آنجا که متعلق به خود کلاس است ، متغیرهای کلاس در تمام اشیا کلاس به اشتراک گذاشته می شود
# بنابراین آنها به طور کلی برای هر شی یک مقدار خواهند داشت 
#  محل تعرسف:  زیر کلاس و قبل از متد سازنده و سایر متدها
#  
```

```python
# نیازی نیست پارامترها با نام ویژگی ها یکسان باشند
class Time(object):
        time_number=0 # متغیر کلاس - Class Variable
        
        def __init__(self,h ,m,s):
            self.hour = h  #  متغیر شی _ instance variable
            self.minute = m
            self.second = s
            Time.time_number += 1 
            # هر دفعه یک نمونه شیء درست می شود تابع سازنده اجرا می شود و به مقدار صفت
            # یک عدد اضافه می شود تا نشانگر تعداد شیء های ساخته شده باشد Time.time_number 
def print_time(t):
    print (t.hour,":", t.minute,":", t.second)            
```

```python
start = Time(9,45,30)
stop = Time(9,45,30)

print_time(start)
print(f" nubmer of Time object = {Time.time_number} ")
```
