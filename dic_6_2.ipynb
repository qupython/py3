```python
#  چالش پیدا کردن " سن " یک فرد خاص
# راه اول با دو لیست 
x1=[10,20,30]
x2=["ali","reza",'Hossein']

for n in range(0,len(x1)):
    if x2[n]=='Hossein':
       print(x1[n])
```

```python
# راه دوم با  یک لیست ترکیبی
x1=["ali",10,"reza",20,'Hossein',30]
for n in range(0,len(x1)):
    if x1[n]=='Hossein':
       print(x1[n+1])
```

```python
# برای حل این مشکل متدی به نام
# dict.get(key,default)
# وجود دارد که کلید یک دیکشنری را گرفته و مقدار آن را برمی گرداند اگر کلید وجود نداشت مقدار
# None
#  بر می گرداند و تولید خطا و توقف در اجرای برنامه نمی نماید
```

```python
student = {
  "name" : "ali",
  "student_number" : "400120005",
  "education" : "industrial engineeringh"
}

print(student.get('age'))

print(student.get('age',18)) # default
```

```python
# setdefault
# مقدار کلید مشخص شده را برمی گرداند. اگر کلید وجود نداشته باشد کلید را با مقدار مشخص شده درج میکند
# dictionary.setdefault(keyname, value)
s=student.setdefault('age',30)
print(s)
student.setdefault('code',43000)
print(student)
```

#  Nested Dictionary --- دیکشنری تو در تو 
#  دو گونه تعریف مستقیم و جداگانه 

```python
# تعریف با استفاده از دیکشنری های میانی به صورت جداگانه
# کلید اصلی را شماره دانشجویی یا کد ملی که مقادیر یکتا هستند می توان انتخاب نمود 
# primary key == Student_Number

student1 = {
  "name" : "mohhamad",
  "score" : 20
}
student2 = {
  "name" : "ali",
  "score" : 19
}
student3 = {
  "name" : "hasan",
  "score" : 18
}

class1 = {
  "4001" : student1,  # member1
  "4002" : student2,  
  "4003" : student3
}

# تعریف مستقیم 
class2 = {"4001":{ "name":"ahmad","score":20},"4002":{"name":"saber","score":19},"4003":{"name":"reza","score": 18}}
```

```python
# برای دسترسی به آیتم‌های یک دیکشنری تودرتو، از نام دیکشنری استفاده کنید،که با دیکشنری بیرونی شروع می شود


print(class1['4001'])

print(class1['4001']['name'])
print(class1['4002']['score'])

print(class1.get('4001'))

n=class1.get('4001').get('name')
print(n)
print(class1)
```
