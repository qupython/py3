```python

import turtle as t

def square(t):          # t  پارامتر
    for i in range(4):
        t.forward(120) 
        t.right(90)
        
        
laki = t.Turtle() # ساخت یک شی از نوع لاک پشت ترسیم کننده هندسی پایتون
square(laki)      # ارسال آرگومان به تابع ترسیم
```

```python
from jupyturtle import make_turtle, forward ,left

def polygon(n, length):
    angle = 360 / n
    for i in range(n):
        forward(length)
        left(angle)
```

```python
# مثال  گرفتن ورودی از کاربر
def stu_name():
    print("enter your name:")
    name = input()
    print("enter your student namber:")
    stu_num = input()
    return name,stu_num
```

```python
# مثال  گرفتن ورودی از کاربر
def stu_score():
    name_score = ['math', 'phy', 'chemis', 'program','quran', 'sport', 'language'] 
    score=[]
    for i in range(0,len(name_score)):
        print("enter your Score:",name_score[i])
        num = int(input())
        score.append(num)
    
    return score
```

```python
# تابع معدل 
def stu_mean(grade,weight):
        sum_num=z=0
        for i in range(len(grade)):
            sum_num += grade[i]*weight[i]
        mean_grade=sum_num/(sum(weight))
         
        return mean_grade

```

```python
# تابع واریانس
def stu_variance(grade):
        sum_num=z=0
        for i in range(len(grade)):
            sum_num += grade[i]
        mean_grade=sum_num/len(grade)

        for x in grade:
            z += (x-mean_grade)**2
        variance=(z/len(grade))*(1/2)
                    
        return variance

```

```python
# -----------------------تابع ارزشیابی معدل
def stu_rank(mean):
    if mean > 16:
        return 'good'
    elif mean > 12:
        return 'medium'
    elif mean > 10:
        return 'Pass'
    else:
        return 'refuse'
```

```python
#------------------------ تابع نمایش نمودار نتایج 
def stu_plot(x,y):
    import matplotlib.pyplot as plt # وارد کردن ماژول پلات

    plt.bar(x, y,color = "blue" , width = 0.1) # رسم نمودار میله ای
    plt.show() # نمایش نمودار
```
