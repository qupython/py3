```python
from jupyturtle import make_turtle, forward ,left

def polygon(n, length):
    angle = 360 / n
    for i in range(n):
        forward(length)
        left(angle)
```

```python
#      
def stu_name():
    print("enter your name:")
    name = input()
    print("enter your student namber:")
    stu_num = input()
    return name,stu_num
```

```python
#      
def stu_score():
    name_score = ['math', 'phy', 'chemis', 'program','quran', 'sport', 'language'] 
    score=[]
    for i in range(0,len(name_score)):
        print("enter your Score:",name_score[i])
        num = int(input())
        score.append(num)
    
    return score
```

```python
#   
def stu_mean(grade,weight):
        sum_num=z=0
        for i in range(len(grade)):
            sum_num += grade[i]*weight[i]
        mean_grade=sum_num/(sum(weight))
         
        return mean_grade

```

```python
#  
def stu_variance(grade):
        sum_num=z=0
        for i in range(len(grade)):
            sum_num += grade[i]
        mean_grade=sum_num/len(grade)

        for x in grade:
            z += (x-mean_grade)**2
        variance=(z/len(grade))*(1/2)
                    
        return variance

```

```python
# -----------------------  
def stu_rank(mean):
    if mean > 16:
        return 'good'
    elif mean > 12:
        return 'medium'
    elif mean > 10:
        return 'Pass'
    else:
        return 'refuse'
```

```python
#------------------------     
def stu_plot(x,y):
    import matplotlib.pyplot as plt #    

    plt.bar(x, y,color = "blue" , width = 0.1) #    
    plt.show() #  
```
