# 3_ آشنایی با متد 



```python
# تعریف کلاس
class Point():
    pass

# تعریف تابع بیرون کلاس
def print_point(p):
    print (p.x, p.y)


blank = Point()
blank.x = 3.0
blank.y = 4.0

print_point(blank)
```

```python
# تعریف تابع داخل کلاس به عنوان یک متد
class Point():
    def print_point(self):      # تعریف تابع داخل کلاس به عنوان متد 
        print (self.x, self.y)  # نماینده و جانشین همه اشیا آن کلاس است  self
        
blank = Point()
blank.x = 3.0
blank.y = 4.0

blank.print_point() 
```

# کپی کردن اشیا و ایجاد شی جدید 

```python
p = Point()
p.x = 3.0
p.y = 4.0
# تعریف شی جدید از شی اول
p1 = p

p1 == p
# هر دو یک شیء هستند
```

Copying
کپی کردن اشیا به صورت یک شیئ جدید

```python

import copy
p2 = copy.copy(p)

print_point(p2)
# دارای داده های یکسان هستند اما به دو شیئ متفاوت ارجاع می کنند
p == p2

# با اینکه مقدارها مساوی هستند نتیجه تست برابری غلط است
```

```python
p is p2
```

```python
class Rectangle():
    pass

rect = Rectangle()
# صفات
rect.width = 10
rect.height = 20
rect.corner=Point()  # شیئ نهفته
rect.corner.x = 0
rect.corner.y = 0
```

```python

rect1 = copy.copy(rect)
rect1 is rect
# شی رک یک کپی شده است و با هم دو شی متفاوت هستند 
```

```python
rect1.corner is rect.corner
# شامل شیئ نهفته کپی نشده است 
```

deep copy
نه تنها خود شیئ بلکه تمام اشیای نهفته در شی و اشیاء نهفته در آنها را نیز کپی می نماید

```python
rect2 = copy.deepcopy(rect)
rect2 is rect
```

```python
rect2.corner is rect.corner
# شیئ نهفته نیز کپی شده و فرق دارد
```
