```python
# ------عملگرهای مقایسه
x = input("enter x:")
x = int(x)
print(f" 10 == {x} :  " , x == 10) # مساوی
print(f" 10 != {x} : " , x != 10) # مخالف
print(f" 10 > {x}  : " , x > 10) # بزرگ تر
print(f" 10 >= {x} : " , x >= 10) # بزرگ تر یا مساوی
print(f" 10 < {x}  : " , x < 10) # کوچک تر
print(f" 10 <= {x} : " , x <= 10) # کوچک تر یا مساوی

# علامتهای <= >= غیرمجاز هستند
# علامت = و == با هم فرق دارند = عملگر انتساب و == عملگر مقایسه تساوی دو شی یا مقدار است 
```

```python
#     تعریف متغیرهای سمبلیک:

import sympy as sp
x = sp.Symbol("x")
y = sp.Symbol("y")
z = sp.Symbol("z")
print(x + y + z - 2 * x - 3 * y)

print(type(x))
```

```python
#    ساده سازی عبارات جبری:

import sympy as sp
x = sp.Symbol("x")
y = sp.Symbol("y")
print(sp.simplify((x*2 + x * y) / x))
```
