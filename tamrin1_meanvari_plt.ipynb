```python
import math

y=x=zigma=0

list_scores = [12,14,16,17,20,15,13] #  لیست نمرات ترم 1

for x in list_scores:
    y=y+x # y +=x
print('sum = ' ,y)

mean=y/len(list_scores)
print('mean =',round(mean, 2))


for x in list_scores:
    zigma=zigma+(x-mean)**2        # مجموع مربعات تفاضل

variance=(zigma/len(list_scores))**(1/2) # محاسبه واریانس
#varian=math.sqrt(zig/len(lis))

print('variance =',round(variance, 2))

print('Result:')
# بررسی شرط ها برای تصمیم گیری نمایش نوع معدل
if mean > 16:
    print('good')
elif mean > 12:
    print('medium')
elif mean > 10:
    print('Pass')
else:
    print('refuse')
    

```

```python
import matplotlib.pyplot as plt # وارد کردن ماژول پلات

#-----------Plot---------
x = ['math', 'phy', 'chemis', 'program','quran', 'sport', 'language'] # لیست نام دروس
# x = [1,2,3,4,5,6,7]

y = list_scores                 # لیست ترم 1
y2 = [13,17,16,18,19,15,10]     # لیست نمره ترم 2
means=[mean]*8                  # ایجاد یک لیست از نمره میانگین

plt.plot(x, y,color = 'red',label = "term_1",marker = '*')  # رسم نمودار 2 بعدی
plt.plot(x,y2,color = 'hotpink' ,label = "term_2")         # رسم نمودار 2 بعدی
plt.plot(means,linestyle = 'dotted',label = "mean_1")

font1 = {'family':'serif','color':'blue','size':18} # تعریف دیکشنری فونت 
font2 = {'family':'serif','color':'darkred','size':15}

plt.title(" Student Results", fontdict = font1,loc = 'center') # تعریف عنوان
plt.xlabel("lessons", fontdict = font2) # تعریف برچسب افقی
plt.ylabel("scores", fontdict = font2)
plt.legend(title = "Legend Title") # تعریف شرح 

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5) 

plt.show() # نمایش نمودار
```
