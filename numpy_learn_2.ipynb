#  (Slicing)تکه برداری 
#   با علامت کولن : که در پایتون معادل با «تا» است و یک بازه "باز" انتخاب می کند 
# [start:end]

# با گام
# [start:end:step]

# اگر سمت چپ علامت : را خالی باشد، یعنی از اندیس صفر شروع شود 
# و اگر سمت راست آن را خالی بگذاریم یعنی تا اندیس آخر ادامه

# علامت : به معنی تمامی عناصر آن بعد است از اول تا آخر


```python
arr = np.array([21,22, 23, 24, 25, 26, 27])
```

```python
# از اندیس 1 تا اندیس 5
print(arr[1:5]) 
```

```python
# از اندیس 4 تا آخر
print(arr[4:]) 
```

```python
# از اندیس صفر تا اندیس 4
print(arr[:4]) 
```

```python
# از اندیس 1 تا اندیس 5 با گام دو
print(arr[1:5:2]) 
```

```python
# از ابتدا تا انتها با گام دو

print(arr[::2]) 
```

#  جبر خطی  
#
تعریف ماتریس2*2 با نامپای


```python
import numpy as np 
A = np.array([[6, 1, 1], 
              [4, -2, 5], 
              [2, 8, 7]]) 
```

```python
print(A.ndim)
```

```python
print(A.shape)
```

محاسبه رنک ماتریس

```python
print("Rank of A:", np.linalg.matrix_rank(A)) 
```

اثر ماتریس = مجموع عناصر قطر اصلی

```python
print("\n Trace of A:", np.trace(A)) 
```

محاسبه مقدار ویژه و بردارهای ویژه

```python
print("\n eigenvalues of A:", np.linalg.eig(A))
```

```python
محاسبه نرم ماتریس 
```

```python
print(A[0:1])
print("\n norm of A:", np.linalg.norm(A[0:1]))
```

محاسبه دترمینان ماتریس

```python
print("\nDeterminant of A:", np.linalg.det(A)) 
```

محاسبه ماتریس معکوس

```python
print("\nInverse of A:\n", np.linalg.inv(A)) 
```

ماتریس به توان 3

```python
print("\nMatrix A raised to power 3:\n", np.linalg.matrix_power(A, 3))
```

y=A.x حل معادلات چند معادله چند مجهول 

```python
y = np.array([[1], [2],[3]])
x=np.linalg.solve(A, y)

print("X solved: \n" , x)
```
