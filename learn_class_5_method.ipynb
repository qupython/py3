```python
# عامل بودن با تابع است 
# یک شیئ بلنک برای پرینت دارید ، print_poin تابع 
blank = Point(5,6)
print_point(blank)
```

```python
# تعریف تابع، داخل کلاس به عنوان یک متد
class Point():
    def __init__(self,x ,y):
        self.x = x
        self.y = y

    def print_point(self):      # تعریف تابع داخل کلاس به عنوان متد 
        print (self.x, self.y)
```

```python
# عاملیت و مسئولیت به شیئ منتقل می شود تا درآینده بتوان توابعی عام ترنوشت و استفاده مجدد ان آسان تر شود
# از طریق شیء فقط به آن دسترسی دارم و مستقلا نمی تواند فراخوانی شود
# خودت را چاپ کن ، blank شیء 

blank = Point(5,6)
blank.print_point()     # print_point(blank)
```

```python
# تعریف کلاس مستطیل با متد سازنده 
# تعریف توابع بیرونی در داخل کلاس به صورت متد
```

```python
class Rectangle():
    def __init__(self,width ,height,corner):
        self.width = width
        self.height = height
        self.corner = Point(corner.x,corner.y)

  
    def find_center(self):
        p = Point(0,0)
        p.x = self.corner.x + self.width/2.0
        p.y = self.corner.y + self.height/2.0
        return p
    
    def grow_rectangle(self, dwidth, dheight):
        self.width += dwidth
        self.height += dheight
        
```

```python
#  نکته* : میتوان پارامتر ورودی داشت که جزو صفات شیء نباشند و برای مثلا تعداد کاری درون یک متد باشد* 
```

```python
rect = Rectangle(100,200,blank)

print (rect.corner.x , rect.corner.y)
```

```python
# متد پیدا کردن مرکز
```

```python
center = rect.find_center()       # center = find_center(rect)

print_point(center)
```

```python
# متد افزایش طول و عرض مستطیل
```

```python
print (rect.height)

rect.grow_rectangle(50, 100)     # grow_rectangle(rect, 50, 100)

print (rect.height)
```

```python
# تعریف تابع،  داخل کلاس به عنوان یک متد
class Time(object):
    def __init__(self,hour ,minute,second):
            self.hour = hour
            self.minute = minute
            self.second = second

    def print_time(self):
        print (self.hour,":", self.minute,":", self.second)
```

```python
# عاملیت و مسئولیت به شیئ منتقل می شود 
# شیئ استارت خودت را چاپ کن
start = Time(9,45,30)

start.print_time()
```
