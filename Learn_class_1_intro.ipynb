```python
# تعریف کلاس جدید مستطیل
```

```python
class Rectangle():
    pass

```


# صفت یک شی می تواند خود یک شی از کلاس دیگر باشد
# شیئی که خود صفت یک شیئ دیگر باشد را شیئ نهفته ( جاسازی شده) می نامند   
#  مانند شیء پوینت که یک صفت کورنر برای شیء مستطیل است  


```python
# ساخت یک نمونه
rect = Rectangle()
# صفات
rect.width = 10
rect.height = 20
rect.corner=Point()  # شیئ نهفته
rect.corner.x = 0
rect.corner.y = 0
```

#  تعریف یک تابع که خروجی آن یک شی است 
#  ونمونه ها به عنوان مقادیر بازگشتی


```python
def find_center(rec):
    p = Point()
    p.x = rec.corner.x + rec.width/2.0
    p.y = rec.corner.y + rec.height/2.0
    return p
```

```python
center = find_center(rect)
print_point(center)
```

# اشیا قابل تغییر هستند


```python
print (rect.width)

rect.width = rect.width + 50
rect.height = rect.width + 100

print (rect.width)
```

# تعریف یک تابع برای تغییر صفات 

```python
def grow_rectangle(rec, dwidth, dheight):
    rec.width += dwidth
    rec.height += dheight
```

```python
print (rect.height)

grow_rectangle(rect, 50, 100)

print (rect.height)
```
