```python
import math
import matplotlib.pyplot as plt

y=x=zigma=0

listn = [12,14,16,17,19,15,13]

for x in listn:
    y=y+x # y +=x
print('sum = ' ,y)

mean=y/len(listn)
print('mean =',mean)

for x in listn:
    zigma=zigma+(x-mean)**2
variance=(zigma/len(listn))**(1/2)
#varian=math.sqrt(zig/len(lis))
print('variance =',variance)

print('Result:')

if mean > 16:
    print('good')
elif mean > 12:
    print('medium')
elif mean > 10:
    print('Pass')
else:
    print('refuse')
    

```

```python
#-----------Plot---------

x = [1,2,3,4,5,6,7]
y = listn
y2 = [13,17,16,18,19,15,10]
means=[mean]*8

plt.plot(x, y,color = 'r',marker = '*',label = "term_1") 
plt.plot(x,y2,c = 'hotpink' ,label = "term_2")
plt.plot(means,linestyle = 'dotted',label = "mean")

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title(" Student Results", fontdict = font1,loc = 'center')
plt.xlabel("lessons", fontdict = font2)
plt.ylabel("scores", fontdict = font2)
plt.legend(title = "Legend Title")

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)

plt.show()
```
