```python
#     تعریف متغیرهای سمبلیک:

import sympy as sp
x = sp.Symbol("x")
y = sp.Symbol("y")
z = sp.Symbol("z")
print(x + y + z - 2 * x - 3 * y)

```

```python
#    ساده سازی عبارات جبری:

import sympy as sp
x = sp.Symbol("x")
y = sp.Symbol("y")
print(sp.simplify((x + x * y) / x))
```
